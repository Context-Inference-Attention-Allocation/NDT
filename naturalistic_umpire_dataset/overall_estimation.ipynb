```python
from phase_2_estimation_ import EMAlgorithm
import numpy as np
import pandas as pd
```

```python
#Load dataframe
filename = 'Umpire_2/utility_estimation_umpire_2.csv'
df = pd.read_csv(filename)

#pre-processing data
df = df[(df['error_in_decision']=='incorrect')|(df['error_in_decision']=='correct')]
df = df.reset_index(drop=True)
df.loc[((df['error_in_decision'] == 'incorrect')),'error_in_decision'] = 0
df.loc[((df['error_in_decision'] == 'correct')),'error_in_decision'] = 1
df['error_in_decision'] = df['error_in_decision'].astype(int)
df.dropna(inplace=True, axis=0, how='any')
print("Number of datapoints: ",len(df))

em_model = EMAlgorithm(df,plot_flag=0,plot_threshold=0.06, seed_=0)

em_model.run_em_algorithm()

weights = em_model.best_weights
theta = em_model.best_theta
bias = em_model.best_bias

em_model.plot_logistic_probabilities()
em_model.plot_context_probabilities()
```
